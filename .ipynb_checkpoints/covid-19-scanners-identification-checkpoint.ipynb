# I. Image pre-processing & Image pipeline creation
### I.1) Creating image df

```python
disease_types = ['COVID', 'non-COVID']
data_dir = '/kaggle/input/sarscov2-ctscan-dataset/'
full_data_dir = os.path.join(data_dir)

full_data = []
for defects_id, sp in enumerate(disease_types):
# what enumerate does -> defects_id = 0, sp = 'COVID' | defects_id = 1, sp = 'non-COVID'
    for file in os.listdir(os.path.join(full_data_dir, sp)):
        full_data.append(['{}/{}'.format(sp, file), defects_id, sp])
#         print(['{}/{}'.format(sp, file), defects_id, sp])
# final data example list of 3 elements: [COVID/filename.png, 0, 'COVID'] 

full = pd.DataFrame(full_data, columns=['File', 'DiseaseID', 'Disease Type'])
# full.head()

# Shuffle the train data set
SEED = 42
full = full.sample(frac=1, random_state=SEED) 
full.index = np.arange(len(full)) # Reset indices
full.head()
```

## I.2) Image data exploration

```python
# image labels plot
plt.hist(full['DiseaseID'], bins = 3)
plt.xticks([0,1])
plt.title('Frequency Histogram of Species')
plt.figure(figsize=(2, 2))
plt.show()
```

```python
# plotting image samples
def plot_defects(defect_types, rows, cols):
    fig, ax = plt.subplots(rows, cols, figsize=(12, 12))
#  subsetting w condition: train['Disease Type'] == defect_types
    defect_files = full['File'][full['Disease Type'] == defect_types].values
    n = 0
    for i in range(rows):
        for j in range(cols):
            image_path = os.path.join(data_dir, defect_files[n])
            # set x ticks and y ticks for images to be blank
            ax[i, j].set_xticks([])
            ax[i, j].set_yticks([])
            ax[i, j].imshow(cv2.imread(image_path))
            n += 1
            
```

### Display images of covid/non covid lung scans

```python
# Displays first n images of each class from training set
plot_defects('COVID', 5, 5)
```

```python
plot_defects('non-COVID', 5, 5)
```

## I.3) Image resizing

```python
IMAGE_SIZE = 64
def read_image(filepath):
    return cv2.imread(os.path.join(data_dir, filepath)) 
# Loading a color image is the default flag
# Resize image to target size
def resize_image(image, image_size):
    return cv2.resize(image.copy(), image_size, interpolation=cv2.INTER_AREA)
```

### Training Images resizing

```python
# reading and resizing images
X_full = np.zeros((full.shape[0], IMAGE_SIZE, IMAGE_SIZE, 3))
for i, file in tqdm(enumerate(full['File'].values)):
    image = read_image(file)
    if image is not None:
        X_full[i] = resize_image(image, (IMAGE_SIZE, IMAGE_SIZE))
# Normalize the data
X_full = X_full / 255.
print('Full Shape: {}'.format(X_full.shape))
```

### Converting labels to categorical

```python
Y_full = full['DiseaseID'].values
Y_full = to_categorical(Y_full, num_classes=2)
# for use w categorical cross entropy
```

## I.4) Splitting Train and Validation data

```python
BATCH_SIZE = 64

# Split the train and validation sets 
X_train, X_val, Y_train, Y_val = train_test_split(X_full, Y_full, test_size=0.2, random_state=SEED)
```

```python
fig, ax = plt.subplots(1, 3, figsize=(15, 15))
for i in range(3):
    ax[i].set_axis_off()
    ax[i].imshow(X_train[i])
    ax[i].set_title(disease_types[np.argmax(Y_train[i])])
```

# II. Model training
## II.1) Setting training parameters

```python
EPOCHS = 20
SIZE=64
N_ch=3
```

## II.2) Transfer learning and model architecture

```python
def build_densenet():
    densenet = DenseNet121(weights='imagenet', include_top=False)

    input = Input(shape=(SIZE, SIZE, N_ch))
    x = Conv2D(3, (3, 3), padding='same')(input)
    
    x = densenet(x)
    
    x = GlobalAveragePooling2D()(x)
    x = BatchNormalization()(x)
    x = Dropout(0.5)(x)
    x = Dense(256, activation='relu')(x)
    x = BatchNormalization()(x)
    x = Dropout(0.5)(x)

    # multi output
    output = Dense(2,activation = 'softmax', name='root')(x)
 

    # model
    model = Model(input,output)
    
    optimizer = Adam(lr=0.002, beta_1=0.9, beta_2=0.999, epsilon=0.1, decay=0.0)
    model.compile(loss='categorical_crossentropy', optimizer=optimizer, metrics=['accuracy'])
    model.summary()
    
    return model
```

```python
model = build_densenet()
annealer = ReduceLROnPlateau(monitor='val_accuracy', factor=0.5, patience=5, verbose=1, min_lr=1e-3)
checkpoint = ModelCheckpoint('model.h5', verbose=1, save_best_only=True)
# Generates batches of image data with data augmentation
datagen = ImageDataGenerator(rotation_range=360, # Degree range for random rotations
                        width_shift_range=0.2, # Range for random horizontal shifts
                        height_shift_range=0.2, # Range for random vertical shifts
                        zoom_range=0.2, # Range for random zoom
                        horizontal_flip=True, # Randomly flip inputs horizontally
                        vertical_flip=True) # Randomly flip inputs vertically

datagen.fit(X_train)
# Fits the model on batches with real-time data augmentation
hist = model.fit_generator(datagen.flow(X_train, Y_train, batch_size=BATCH_SIZE),
               steps_per_epoch=X_train.shape[0] // BATCH_SIZE,
               epochs=EPOCHS,
               verbose=2,
               callbacks=[annealer, checkpoint],
               validation_data=(X_val, Y_val))
```

```python
# model architechture
plot_model(model, to_file='convnet.png', show_shapes=True,show_layer_names=True)
Image(filename='convnet.png') 
```

# III. Model Evaluation
## III.1) final loss and accuracy

```python
#model = load_model('../output/kaggle/working/model.h5')
final_loss, final_accuracy = model.evaluate(X_val, Y_val)
print('Final Loss: {}, Final Accuracy: {}'.format(final_loss, final_accuracy))
```

## III.2) Confusion matrix

```python
Y_pred = model.predict(X_val)

Y_pred = np.argmax(Y_pred, axis=1)
Y_true = np.argmax(Y_val, axis=1)

cm = confusion_matrix(Y_true, Y_pred)
plt.figure(figsize=(12, 12))
ax = sns.heatmap(cm, cmap=plt.cm.Greens, annot=True, square=True, xticklabels=disease_types, yticklabels=disease_types)
ax.set_ylabel('Actual', fontsize=40)
ax.set_xlabel('Predicted', fontsize=40)
```

## III.3) Accuracy and loss curve

```python
# accuracy plot 
plt.plot(hist.history['accuracy'])
plt.plot(hist.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()

# loss plot
plt.plot(hist.history['loss'])
plt.plot(hist.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```
